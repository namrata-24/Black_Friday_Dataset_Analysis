## Data Cleaning

```python
data.isna().any()
```

```python
data2 = data.copy()
```

```python
data2.head()
```

```python
data2.fillna(value=0,inplace = True)
```

```python
data2.isna().any()
```

```python
data2.head()
```

```python
data2.drop('Product_ID' , axis = 1,inplace = True)
```

```python
for i in range(101,len(data2)):
    if data2.iloc[i]['Stay_In_Current_City_Years'] == '4+':
        data2.Stay_In_Current_City_Years.iloc[i]= 5
```

```python
data2.drop(columns =["User_ID"],inplace = True,axis = 1)
```

```python
sb.barplot(x = data2['City_Category'],y = data2['Purchase'])
```

```python
def fun(num):
    if num == "C":
        return 2
    elif num == 'A':
        return 1;
    elif num == 'B':
        return 1.5

data2['City_Category'] = data2['City_Category'].apply(fun)
```

```python
data2.head()
```

```python
def fun1(num):
    if num == '4+':
        return 5
    else:
        return num 

data2['Stay_In_Current_City_Years'] = data2['Stay_In_Current_City_Years'].apply(fun1)
```

```python
data2.head()
```

```python
from sklearn.model_selection import train_test_split
features = data2[['Occupation','City_Category','Stay_In_Current_City_Years','Marital_Status','Product_Category_1','Product_Category_2','Product_Category_3']]
output = data2['Purchase']
train_X ,train_Y,test_X,test_Y = train_test_split(features,output,test_size=0.2)
```
